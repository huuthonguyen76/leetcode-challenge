```python
import pandas as pd

# Create test data with 10 rows: Employee, Region, Salary
employee_data = {
    "Employee": ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
    "Region": ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    "Salary": [50000, None, 55000, 62000, 58000, 54000, 61000, None, 59000, 60500]
}

df_emp = pd.DataFrame(employee_data)
print(df_emp)

df_region_group = df_emp.groupby(['Region'])['Salary'].mean()
print(df_region_group)

df_emp['Salary'] = df_emp.apply(
    lambda row: row['Salary'] if row['Salary'] is pd.notnull(row['Salary']) else df_region_group[row['Region']], axis=1
)
print(df_emp)
```
