```python
import torch

# input, output are tensors on the GPU
def solve(input: torch.Tensor, output: torch.Tensor, N: int):
    input[input < 0] = input[input < 0] * 0.01
    output.copy_(input)

x = torch.tensor([-1, 0, 1, 2], dtype=torch.float32)
y = torch.zeros_like(x)


solve(x, y, 4)
print(y)
```
