```python
import torch

# image is a tensor on the GPU
def solve(image: torch.Tensor, width: int, height: int):
    original_shape = image.shape
    image_new = image.view(-1, 4).clone()
    image_new[:, :3] = 255 - image_new[:, :3]
    image_new = image_new.view(original_shape)
    image = image_new
    image.copy_(image_new)
    


image = torch.tensor([255, 0, 128, 255, 0, 255, 0, 255])
width = 1
height = 2

solve(image, width, height)

print(image)
```
