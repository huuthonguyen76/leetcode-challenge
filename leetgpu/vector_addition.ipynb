```python
import torch

# A, B, C are tensors on the GPU
def solve(A: torch.Tensor, B: torch.Tensor, C: torch.Tensor, N: int):
    C.add_(A)
    C.add_(B)

ts_1 = torch.randn(5, dtype=torch.float32)
ts_2 = torch.randn(5, dtype=torch.float32)
ts_3 = torch.tensor([0] * 5, dtype=torch.float32)

solve(ts_1, ts_2, ts_3, 10)

```
