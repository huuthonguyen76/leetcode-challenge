```python
import torch


# input, output are tensors on the GPU
def solve(input: torch.Tensor, output: torch.Tensor, N: int, M: int, K: int):
    output[0] = (input == K).sum()

ts_1 = torch.tensor([
    [1, 2, 3],
    [4, 2, 1]
])

ts_2 = torch.zeros_like(ts_1)

solve(ts_1, ts_2, 1, 1, 2)

print(ts_2)
```

```python
import torch
import math


def sigmoid(x):
    return 1.0 / (1 + (math.e ** -x))

# input, output are tensors on the GPU
def solve(input: torch.Tensor, output: torch.Tensor, N: int):
    input = input * sigmoid(input)
    output.copy_(input)
    print(input)

ts_1 = torch.tensor([0.5, 1.0, -0.5])

solve(ts_1, ts_1, 3)
```
